```python
import sqlite3

# Veritabanı bağlantısını oluştur
conn = sqlite3.connect('ilk_veritabanı.db')
cursor = conn.cursor()

# Eser tablosunu oluştur
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Eser (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        ad TEXT NOT NULL,
        yazar_id INTEGER,
        yayinevi_id INTEGER,
        FOREIGN KEY (yazar_id) REFERENCES Yazar(id),
        FOREIGN KEY (yayinevi_id) REFERENCES Yayinevi(id)
    )
''')

# Konu Basliklari tablosunu oluştur
cursor.execute('''
    CREATE TABLE IF NOT EXISTS "Konu Basliklari" (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        konu_basligi TEXT NOT NULL
    )
''')

# Yayinevi tablosunu oluştur
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Yayinevi (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        yayinevi_adi TEXT NOT NULL
    )
''')

# Yazar tablosunu oluştur
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Yazar (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        yazar_adi TEXT NOT NULL
    )
''')

# Veritabanı bağlantısını kapat
conn.close()

import sqlite3

class VeritabaniUygulamasi:
    def __init__(self, veritabani_adi):
        self.conn = sqlite3.connect(veritabani_adi)
        self.cursor = self.conn.cursor()

    def eserleri_listele(self):
        self.cursor.execute('SELECT * FROM Eser')
        eserler = self.cursor.fetchall()
        for eser in eserler:
            print(eser)

    def eser_ekle(self, ad, yazar_id, yayinevi_id):
        self.cursor.execute('''
            INSERT INTO Eser (ad, yazar_id, yayinevi_id)
            VALUES (?, ?, ?)
        ''', (ad, yazar_id, yayinevi_id))
        self.conn.commit()
        print('Eser eklendi.')

    def konu_basliklari_listele(self):
        self.cursor.execute('SELECT * FROM "Konu Basliklari"')
        konu_basliklari = self.cursor.fetchall()
        for konu in konu_basliklari:
            print(konu)

    def yayinevleri_listele(self):
        self.cursor.execute('SELECT * FROM Yayinevi')
        yayinevleri = self.cursor.fetchall()
        for yayinevi in yayinevleri:
            print(yayinevi)

    def yazarlari_listele(self):
        self.cursor.execute('SELECT * FROM Yazar')
        yazarlar = self.cursor.fetchall()
        for yazar in yazarlar:
            print(yazar)

    def kapat(self):
        self.conn.close()

# Örnek kullanım
uygulama = VeritabaniUygulamasi('ilk_veritabanı.db')

# Eserleri listele
print("Eserleri Listele:")
uygulama.eserleri_listele()

# Konu başlıklarını listele
print("\nKonu Başlıkları Listele:")
uygulama.konu_basliklari_listele()

# Yayınevlerini listele
print("\nYayınevleri Listele:")
uygulama.yayinevleri_listele()

# Yazarları listele
print("\nYazarları Listele:")
uygulama.yazarlari_listele()

# Yeni eser ekle
print("\nYeni Eser Ekle:")
uygulama.eser_ekle('Eser1', 1, 1)

# Eserleri listele
print("\nEserleri Listele:")
uygulama.eserleri_listele()

# Uygulama kapat
uygulama.kapat()
```
