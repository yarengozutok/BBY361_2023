```python
import sqlite3

# Veritabanı bağlantısı oluştur
conn = sqlite3.connect('ilk_veritabanı.db')
cursor = conn.cursor()

# Eserler tablosunu oluştur
cursor.execute('''
    CREATE TABLE IF NOT EXISTS ilk_veritabani (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        ad TEXT NOT NULL,
        yazar TEXT NOT NULL,
        tarih TEXT NOT NULL,
        tur TEXT NOT NULL
    )
''')

# Veritabanı bağlantısını kapat
conn.close()

class EserUygulamasi:
    def __init__(self, veritabani_adi):
        self.conn = sqlite3.connect(veritabani_adi)
        self.cursor = self.conn.cursor()

    def eserleri_listele(self):
        self.cursor.execute('SELECT * FROM eserler')
        eserler = self.cursor.fetchall()
        for eser in eserler:
            print(eser)

    def eser_ekle(self, ad, yazar, tarih, tur):
        self.cursor.execute('''
            INSERT INTO eserler (ad, yazar, tarih, tur)
            VALUES (?, ?, ?, ?)
        ''', (ad, yazar, tarih, tur))
        self.conn.commit()
        print('Eser eklendi.')

    def eser_ara(self, kelime):
        self.cursor.execute('''
            SELECT * FROM eserler
            WHERE ad LIKE ? OR yazar LIKE ? OR tarih LIKE ? OR tur LIKE ?
        ''', (f'%{kelime}%', f'%{kelime}%', f'%{kelime}%', f'%{kelime}%'))
        eserler = self.cursor.fetchall()
        for eser in eserler:
            print(eser)

    # Bonus: Eser güncelleme
    def eser_guncelle(self, eser_id, ad, yazar, tarih, tur):
        self.cursor.execute('''
            UPDATE eserler
            SET ad=?, yazar=?, tarih=?, tur=?
            WHERE id=?
        ''', (ad, yazar, tarih, tur, eser_id))
        self.conn.commit()
        print('Eser güncellendi.')

    # Bonus: Eser silme
    def eser_sil(self, eser_id):
        self.cursor.execute('DELETE FROM eserler WHERE id=?', (eser_id,))
        self.conn.commit()
        print('Eser silindi.')

    def kapat(self):
        self.conn.close()

# Örnek kullanım
uygulama = EserUygulamasi('eserler.db')

# Eserleri listele
print("Eserleri Listele:")
uygulama.eserleri_listele()

# Yeni eser ekle
print("\nYeni Eser Ekle:")
uygulama.eser_ekle('Eser1', 'Yazar1', '2022-01-01', 'Roman')

# Eserleri listele
print("\nEserleri Listele:")
uygulama.eserleri_listele()

# Eser ara
print("\nEser Ara:")
uygulama.eser_ara('Eser1')

# Bonus: Eser güncelle
print("\nEser Güncelle:")
uygulama.eser_guncelle(1, 'Yeni Eser Adı', 'Yeni Yazar', '2023-01-01', 'Şiir')
uygulama.eserleri_listele()

# Bonus: Eser sil
print("\nEser Sil:")
uygulama.eser_sil(1)
uygulama.eserleri_listele()

# Uygulama kapat
uygulama.kapat()
```
